### 1. Enviornment setup
#### Script languages
Scripts were developed using R, python 2.7x, and BASH scripting languages.

Follow the instruction on [CHROMATIX](https://bitbucket.org/aperezrathke/chr-loc-mbi-mam/src/master/) to install the required packages

#### Directory structure
We assume a working directory [WorkDir] which is freely named by the user. However, for high performance computing (HPC) scripts running on the UIC Extreme HPC cluster, the [WorkDir] path is assumed to be ~/chr-working.

Via command line, navigate to [WorkDir] and git clone the CHR-FOLDER and CMX projects within this directory:

```bash
git clone https://bitbucket.org/aperezrathke/chr-folder.git folder
```
This will copy the source code for the 3-D chromatin folder into the [WorkDir]/folder subdirectory.

```bash
git clone https://bitbucket.org/aperezrathke/cmx.git
```

This will copy the source code for the Bayesian Hi-C single-cell deconvolver into the [WorkDir]/cmx subdirectory.
Next, still from within [WorkDir], clone the MBI-MAM project:

```bash
git clone https://bitbucket.org/aperezrathke/chr-loc-mbi-mam.git ./loci/mbi-mam
```

This will copy the scripts for many-body analysis of the GM12878 cell line into the [WorkDir]/loci/mbi-mam subdirectory.
We will refer to the local path [WorkDir]/loci/mbi-mam as [RepoDir].

##### Output directory

Finally, within [RepoDir], many scripts will generate output data to the directory [RepoDir]/gm12878/fractal2d.hic5k.tad2mb/output. This folder is generated automatically when needed. All script output is stored within this directory.



### 2. Baking

Navigate to the GM12878 common subdirectory at [RepoDir]/gm12878/fractal2d.hic5kb.tad2mb/common.

From within a command line terminal, run the bake.py script. This will work for most standard linux environments such as Ubuntu or CentOs.

### 3. Null distribution
The next step is to generate a uniform random chromatin 3-D polymer ensemble. This 3-D polymer ensemble will be repeatedly bootstrapped to generate a null distribution over chromatin contact frequencies. Using this bootstrapped distribution, we can then call non-random (i.e. significant or not well explained through uniform random polymer folding) chromatin interactions within the Hi-C.

- 1. Navigate to the GM12878 null subdirectory at [RepoDir]/gm12878/fractal2d.hic5kb.tad2mb/null.
- 2. Run the script `null_folder_run.py`. This will generate a large ensemble of uniform random 3-D chromatin polymers.
- 3. Run the script `null_blb_run_parallel.sh` to generate the bootstrap null distribution over chromatin contact frequencies. Note, this script simply spawns multiple instances of the `null_blb_run.py` script to generate multiple Bag of Little Bootstraps (BLB) replicates.
- 4. Run the script `null_cache_run.sh`. This will convert the output from the previous steps into RData format usable for R analysis.

### Calling non-random interactions
We can now use the bootstrap null distribution to call non-random interactions within the Hi-C data. However, first we must identify the TAD regions of interest and pull down the corresponding Hi-C data from Rao et al.

All scripts will be located in the the GM12878 common subdirectory at [RepoDir]/gm12878/fractal2d.hic5kb.tad2mb/common.

### Identify and format loci for analyzing

Prepare the table of loci you are interested in, in the format of 

| name | chr | bp_start_tad_med | bp_end_tad_med |
|-----------------|-----------------|-----------------|-----------------|
|chr2.230805000.231690000|chr2|230805000|231690000|
| ...	| ...	| ...	| ... |



Run the script `tad_juicer.py`. This will process the table of loci we are interested in and process 4DN downloaded Hi-C data using Juicer.

### Clean TADs

Now that we have downloaded the Hi-C and generated the null distribution, we can now call the non-random interactions within the Hi-C. We call this process cleaning the data - i.e. retaining only the non-random interactions for subsequent downstream analysis.

Run the script `tad_clean_run.sh`. This will call non-random interactions at each TAD listed within the master table of TAD regions. It will also generate corresponding heat maps.
